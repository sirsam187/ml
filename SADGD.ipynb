```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data set.
X = np.array([0.5, 2.5])
Y = np.array([0.2, 0.9])

# Initialize the neural network with random weights.
w = np.random.randn()
b = np.random.randn()

# Calculate the output of the neural network.
y_pred = w * X + b

# Calculate the squared error loss.
loss = np.mean((y_pred - Y) ** 2)

# Update the network parameters using batch Adaptive delta gradient descent optimizer function.
learning_rate = 0.01
w_update = learning_rate * np.mean((y_pred - Y) * X)
b_update = learning_rate * np.mean((y_pred - Y))

# Update the weights and bias.
w += w_update
b += b_update

# Display the updated weight and bias values.
print("Updated weight:", w)
print("Updated bias:", b)

# Plot the accuracy w.r.t. epoch values.
epochs = np.arange(1, 100)
accuracies = []
for epoch in epochs:
  # Calculate the output of the neural network.
  y_pred = w * X + b

  # Calculate the squared error loss.
  loss = np.mean((y_pred - Y) ** 2)

  # Calculate the accuracy.
  accuracy = np.mean(y_pred == Y)

  # Add the accuracy to the list.
  accuracies.append(accuracy)

# Plot the accuracy w.r.t. epoch values.
plt.plot(epochs, accuracies)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show()
```
